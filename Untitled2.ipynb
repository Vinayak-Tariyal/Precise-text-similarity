```python

import re
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')


data = pd.read_csv('/content/Precily_Text_Similarity.csv')

# Text preprocessing
# Lowercase conversion
data['text1'] = data['text1'].str.lower()
data['text2'] = data['text2'].str.lower()

# Removing  punctuation
data['text1'] = data['text1'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
data['text2'] = data['text2'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

# Removing  numbers
data['text1'] = data['text1'].apply(lambda x: re.sub(r'\d+', '', x))
data['text2'] = data['text2'].apply(lambda x: re.sub(r'\d+', '', x))

# Removing  stopwords
stopwords_list = set(stopwords.words('english'))
data['text1'] = data['text1'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords_list]))
data['text2'] = data['text2'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords_list]))

# Lemmatization
lemmatizer = WordNetLemmatizer()
data['text1'] = data['text1'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
data['text2'] = data['text2'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))

print(data.head())
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
import faiss

data = pd.read_csv('/content/Precily_Text_Similarity.csv')




train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Loading  the pre-trained Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encoding  train_data and test_data text
train_text_embeddings = model.encode(train_data['text1'].tolist() + train_data['text2'].tolist())
test_text_embeddings = model.encode(test_data['text1'].tolist() + test_data['text2'].tolist())

index = faiss.IndexFlatIP(model.get_sentence_embedding_dimension())
index.add(np.array(train_text_embeddings))

k = 1  # Number of nearest neighbors to retrieve (in this case, the most similar pair)
_, similarity_indices = index.search(np.array(test_text_embeddings), k)

# Extract similarity scores
similarity_scores = []
for i in range(len(similarity_indices)):
    similarity_scores.append(similarity_indices[i][0])

# Split similarity scores back into train and test sets
train_similarity_scores = similarity_scores[:len(train_data)]
test_similarity_scores = similarity_scores[len(train_data):]

print("Train Similarity Scores:", train_similarity_scores)
print("Test Similarity Scores:", test_similarity_scores)
```

```python

import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the trained model
model = SentenceTransformer('bert-base-uncased')

# Function to predict similarity score between two sentences
def predict_similarity(sentence1, sentence2):
    # Encode the sentences
    sentence1_embedding = model.encode([sentence1])[0]
    sentence2_embedding = model.encode([sentence2])[0]

    # Calculate cosine similarity between the embeddings
    similarity_score = cosine_similarity([sentence1_embedding], [sentence2_embedding])[0][0]

    return similarity_score

# Example usage
sentence1 = "I am hungry "
sentence2 = "I am hungry"

similarity_score = predict_similarity(sentence1, sentence2)
print("Similarity Score:", similarity_score)
```
